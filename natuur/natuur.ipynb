```python
import geopandas as gpd
import pandas as pd
import folium
import json
```

```python
stations = gpd.read_file('../data/stations.geojson')
#natuurpunten = gpd.read_file('../data/natuurpunt_v3.geojson')
#plekjes = gpd.read_file('../data/plekjes.geojson')
#plekjes2 = gpd.read_file('../data/plekjes_2.geojson')
plekjes = gpd.read_file('../data/alle_plekjes.geojson')
```

```python
map1 = folium.Map(location=[51.11, 4.46], zoom_start=12)
```

## stations

```python
for index, row in stations.iterrows():
    folium.Circle(
        radius=100,
        location=[row['geometry'].centroid.y, row['geometry'].centroid.x],
        popup=row['name'],
        color="crimson",
        fill="crimson",
    ).add_to(map1)
```

## natuur

```python
for index, row in plekjes.iterrows():
    geo_j = folium.GeoJson(data=row['geometry'],
                           style_function=lambda x: {'lineColor': 'green','fillColor': 'green'})
    folium.Popup(row['name']).add_to(geo_j)

    geo_j.add_to(map1)
```
